```python
import pandas as pd
dataset=pd.read_csv("preplacement.csv")
dataset
```

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
 #independent sample-unpaired T test(Different group(categorical)male,female same condition)
male=dataset[dataset['gender']=='M']['salary']
print(male)
female=dataset[dataset['gender']=='F']['salary']
print(female)
ttest_ind(male,female)
```

```python
#the above result shows a pvalue slightly greater than 0.05 which means there is a very low significant difference
```

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
#dependent sample-paired T test(same group(categorical) different condition)
male=dataset[dataset['gender']=='M']['hsc_p']
print(male)
female=dataset[dataset['gender']=='M']['ssc_p']
print(female)
ttest_ind(male,female)
```

```python
#the above result shows a pvalue greater than 0.05 which means there is no significant difference
```
